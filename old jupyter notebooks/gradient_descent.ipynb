```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.array(range(80))/20
ys = 1.2*(xs-2)*(xs-2)+3.2

plt.plot(xs,ys)
plt.show()
```

```python
def derivative(x):
    return 2*1.2*(x-2)

def update(x, step_size=0.6):
    newx = x - step_size * derivative(x)
    newy = 1.2*(newx-2)*(newx-2)+3.2
    return newx, newy

x = 0.1
y = 1.2*(x-2)*(x-2)+3.2
print(x,y)

plt.plot(xs,ys)
plt.scatter(x,y)
plt.show()
```

```python
x, y = update(x, 0.6)
print(x,y)

plt.plot(xs,ys)
plt.scatter(x,y)
plt.show()
```
