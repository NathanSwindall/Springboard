# Anime News Network Encylopedia API

```python
# To use the Anime News Network Encyclopedia API we will have to know how to use it
# This information is found here: https://www.animenewsnetwork.com/encyclopedia/api.php
```

```python
# request the information
# make sure to try the web address in the browser to see how the xml looks
anime_url = 'https://cdn.animenewsnetwork.com/encyclopedia/reports.xml?id=155&type=anime&nlist=all'
annep = requests.get(anime_url)
```

```python
# Let's explore the request call that we got back
print(annep.headers['content-type'])
print(annep.encoding)
```

```python
# We see that we are getting an xml response back. There are luckily good packages in python that will
# parse this data
# https://stackoverflow.com/questions/18308529/python-requests-package-handling-xml-response

anime_report = ElementTree.fromstring(annep.content)
```

```python
# Here we are going to get the correct tags that we want for information
for anime in anime_report.findall('item'):
    anime_id = anime.find('id').text
    name = anime.find('name').text
    anime_type = anime.find('type').text
    print(anime_id, name,anime_type)
    
```

```python
# Now let's get this information and store it into a panda dataframe
# We will create the panda datafram from a dictionary so we need to make a dictionary first
import pandas as pd



# Let's make a dictionary
anime_dict = {'id': [], 'name': [], 'type': []}
for anime in anime_report.findall('item'):
    anime_id = anime.find('id').text
    anime_name = anime.find('name').text
    anime_type = anime.find('type').text
    
    anime_dict['name'].append(anime_name)
    anime_dict['id'].append(anime_id)
    anime_dict['type'].append(anime_type)
    
anime_df = pd.DataFrame(anime_dict)
```

```python
# Let's search one more file to see what other data we can extract
anime_one_url = 'https://cdn.animenewsnetwork.com/encyclopedia/api.xml?anime=1'
anime_one_re = requests.get(anime_one_url)
anime_one_episode = ElementTree.fromstring(anime_one_re.content)


my_dict = {'id': "",
           "type": "",
           "Genres": [], 
           "Main title": "",
           "Alternative title": [], 
           "Pictures": [],
           "Themes": [],
           "Num_ep": "",
           "Plot": "",
           "votes" : "",
           "weighted_score": "",
           "bayesian_score": "",
           "Vintage": "",
           "Running time": ""}

print(anime_one_episode[0].attrib["id"])

for item in anime_one_episode[0]:
    
    
    
    if 'type' in item.attrib:
        atr = item.attrib['type']
        if atr == "Main title":
            my_dict["Main title"] = item.text
        elif atr == "Picture":
            my_dict["Pictures"].append(item.attrib['src'])
        elif atr == "Alternative title":
            my_dict["Alternative title"].append(item.text)
        elif atr == "Genres":
            my_dict["Genres"].append(item.text)
        elif atr == "Themes":
            my_dict["Themes"].append(item.text)
        elif atr == "Plot Summary":
            my_dict["Plot"] = item.text
        elif atr == "Running time":
            my_dict["Running time"] = item.text
        elif atr == "Vintage":
            my_dict["Vintage"] = item.text
        elif atr == "Number of episodes":
            my_dict["Num_ep"] = item.text
    if 'nb_votes' in item.attrib:
            my_dict['votes'] = item.attrib["nb_votes"]
            my_dict['weighted_score'] = item.attrib["weighted_score"]
            my_dict['bayesian_score'] = item.attrib['bayesian_score']

    
my_dict
```

```python
# We were able to get all the information for 1 anime. Let's make a function that will return a dictionary
# with all the information that we want

def get_anime_info(anime_response):
    my_dict = {
           "id": anime_response[0].attrib["id"],
           "type": anime_response[0].attrib["type"],
           "Genres": [], 
           "Main title": "",
           "Alternative title": [], 
           "Pictures": [],
           "Themes": [],
           "Num_ep": "",
           "Plot": "",
           "votes" : "",
           "weighted_score": "",
           "bayesian_score": "",
           "Vintage": "",
           "Running time": ""}
    
    
    for item in anime_response[0]:
        if 'type' in item.attrib:
            atr = item.attrib['type']
        
            # Get Main Title
            if atr == "Main title":
                my_dict["Main title"] = item.text
        
            # Get the source picture
            elif atr == "Picture":
                my_dict["Pictures"].append(item.attrib['src'])
            
            # Get the alternative titles
            elif atr == "Alternative title":
                my_dict["Alternative title"].append(item.text)
                
            # Get the different Genres
            elif atr == "Genres":
                my_dict["Genres"].append(item.text)
            
            # Get the different themes
            elif atr == "Themes":
                my_dict["Themes"].append(item.text)
            
            # Get the plot summary
            elif atr == "Plot Summary":
                my_dict["Plot"] = item.text
            
            # Get the running time
            elif atr == "Running time":
                my_dict["Running time"] = item.text
            
            # Get the year it ran
            elif atr == "Vintage":
                my_dict["Vintage"] = item.text
            
            # Get the number of episode
            elif atr == "Number of episodes":
                my_dict["Num_ep"] = item.text
            
            # Get the votes
        
        if 'nb_votes' in item.attrib:
            my_dict['votes'] = item.attrib["nb_votes"]
            my_dict['weighted_score'] = item.attrib["weighted_score"]
            if 'bayesian_score' in item.attrib:
                my_dict['bayesian_score'] = item.attrib['bayesian_score']
    
    return my_dict
    
```

```python
# This code will send 100 requests to Anime News Network api
import time

anime_requests = []
anime_url_base = 'https://cdn.animenewsnetwork.com/encyclopedia/api.xml?anime='


for row in hundred_animes.index:
    anime_url_finished = anime_url_base + str(hundred_animes['id'][row])
    anime_req = requests.get(anime_url_finished)
    anime_requests.append(ElementTree.fromstring(anime_req.content))
    time.sleep(1)

len(anime_requests)
```

```python
# lets try to get some of the data into an array and see how long it takes to get it. This will only get 100 entries of 10 anime 
# each so, we will have 1000 entries of anime

# get initial request
kitsu_requests = []
kitsu_start_url = 'https://kitsu.io/api/edge/anime'
kitsu_start_re = requests.get(kitsu_start_url)
kitsu_start_json = kitsu_start_re.json()
kitsu_next = kitsu_start_json['links']['next']


# create a counter
count = 0


# get the other requests
while count < 100:
    kitsu_request = requests.get(kitsu_next)  # get request
    kitsu_request_json = kitsu_request.json()  # turn into a python dictionary
    kitsu_requests.append(kitsu_request_json)  # add data do the kitsu requests
    if 'next' not in kitsu_request_json['links']: # check if there is a next link
        break
    else:
        kitsu_next = kitsu_request_json['links']['next']  # set the next link
        count += 1
    
    
    
```

```python
# Make sure the array that we built has 100 different entries
len(kitsu_requests)
```

```python
# Lets extract all the data from all 100 entries of data
#

import itertools # use this for flattening the list of lists for jsons files


kitsu_data_jsons = []

for kitsu_request in kitsu_requests:
    kitsu_data = kitsu_request['data']
    kitsu_data_jsons.append(kitsu_data)
    
kitsu_data_jsons = list(itertools.chain.from_iterable(kitsu_data_jsons))

print(len(kitsu_data_jsons))
    
```
